# optimize

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0.0,1.0,0.1)
def f(x):
    return -np.exp(-(x-0.7)**2)
plt.plot(x,f(x),'o-')
```

# Scipy Integrate

```python
import scipy.integrate as intg

def integrad(x):
    return x**2
ans, _ = intg.quad(integrad,0,1)
ans
```

# Scipy Linalg

```python
from scipy import linalg
import numpy as np
data = np.array ([[1,2,3],[3,4,5],[5,6,7]])
linalg.det(data)
```
