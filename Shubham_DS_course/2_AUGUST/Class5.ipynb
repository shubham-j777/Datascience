# Merge 

```python
import pandas as pd
import numpy as np
```

```python

leftdf = pd.DataFrame({'Alphabet':['a','b','c','d'],
                      'number':['1','2','3','4'],
                      'alphanumeric':['a1','b2','c3','d4']})
rightdf = pd.DataFrame({'Alphabet':['a','b','c','d'],
                      'number':['4','5','6','7'],
                      'alphanumeric':['e1','f2','g3','h4']})
a = pd.merge(leftdf,rightdf,on = 'Alphabet')
a
```

# JOIN

```python
df1 = pd.DataFrame({'keys':['0','1','2','3'],
                   'A':['A0','A1','A2','A3']})
df2 = pd.DataFrame({'keys':['0','1'],
                   'A':['A0','A1']})
a = df1.join(df2,lsuffix = '_f_df',rsuffix = '_s_df')
a
```

```python
#2df join
df1 = pd.DataFrame({'Choclate_name':['dairy milk','kitkat','perk','Hershey'],
                   'Prices':['20','20','10','10']})
df2 = pd.DataFrame({'Choclate_name':['Lotus'],
                   'Prices':['100']})
a = df1.join(df2,lsuffix = '_f_df',rsuffix = '_s_df')
a
```

```python
#3df join
df1 = pd.DataFrame({'Choclate_name':['dairy milk','kitkat','perk','Hershey'],
                   'Prices':['20','20','10','10']})
df2 = pd.DataFrame({'Choclate_name':['Lotus'],
                   'Prices':['100']})
df3 = pd.DataFrame({'Choclate_name':['Amul'],
                   'Prices':['30']})
#join = 
a1 = df1.join(df2,lsuffix = '_first_df',rsuffix = '_second_df')
b1 = df3.join(a1,lsuffix = '_third_df',rsuffix = '_fourth_df')
b1
```

# Concatenate

```python
first_df = pd.DataFrame({'name':['a','b','c'],
                         'age':[1,2,3]},
                        index = [0,1,2])
second_df = pd.DataFrame({'name':['d','e','f'],
                         'age':[1,2,3]},
                        index = [3,4,5])
third_df = pd.DataFrame({'name':['g','h','i'],
                         'age':[1,2,3]},
                        index = [6,7,8])
join = pd.concat([first_df,second_df,third_df])
join
```

```python
a = np.linspace(0,10,20)
b = np.linspace(10,0,20)
fig = plt.figure(figsize=(6,6))
ax1=plt.subplot(221)
ax2= plt.subplot(222)
ax1.plot(a,linewidth = 1.0,linestyle = '--',color = 'g',alpha = 1,marker = 'o')
ax2.plot(b,linewidth = 1.0,linestyle = '--',color = 'b',alpha = 1,marker = '^')
plt.show()
```

```python
#Q 5 line in 1 graph
a = np.linspace(0,10,5)
b = np.linspace(0,10,8)
c = np.linspace(0,10,9)
d = np.linspace(0,10,11)
e = np.linspace(0,10,16)

ax1 = plt.subplot()
ax1.plot(a,color='g',linestyle = '-',marker ='o')
ax1.plot(b,color='b',linestyle = '-',marker ='o')
ax1.plot(c,color='r',linestyle = '-',marker ='o')
ax1.plot(d,color='black',linestyle = '-',marker ='o')
ax1.plot(e,color='pink',linestyle = '-',marker ='o')
plt.show()
```

#  Bar plot

```python
data = {'apples':10,'oranges':15,'lemons':5,'limes':20}
name = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (15,7))
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax1.bar(name,values,color = ('g','b'),alpha = 0.2)
ax2.barh(name,values)
plt.xlabel('Names')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.grid(True)
plt.savefig('Barplot.jpg')
plt.show
```

# Pie chart

```python
labels = ['frogs','hogs','dogs','logs']
sizes = [15,30,45,10]
explode = (0.1,0,0,0)
fig = plt.figure(figsize = (5,5))
ax1 = plt.subplot()
ax1.pie(sizes,explode=explode,labels=labels,autopct='%1.2f%%',startangle=0,shadow=True)
plt.show()
```
