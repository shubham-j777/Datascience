# voilenplot

```python
t = [20,4,1,30,20,10,20,70,150,10]
o = [10,3,1,15,17,2,30,44,2,1]
d = [30,20,10,5,20,10,60,20,40,1]
da = list([t,o,d])
fig = plt.figure(figsize = (7,7))
a = plt.subplot(121)
b = plt.subplot(122)
a.violinplot(da,showmeans = True ,showmedians  = False,)
b.violinplot(da,showmeans = True,showmedians = True)
plt.show()
```

```python
x_pos = [0,0]
y_pos = [0,0]
x_dircn = [1,0]
y_direcn = [1,-1]
fid = plt.figure(figsize = (5,5))
a = plt.subplot()
a.quiver(x_pos,y_pos,x_dircn,y_direcn,scale = 7)
plt.show()
```

```python
x_pos = [0,0]
y_pos = [0,0]
x_dircn = [1,-1]
x_dir2 =[1,1]
y_direcn = [1,-1]
fid = plt.figure(figsize = (5,5))
a = plt.subplot(121)
b = plt.subplot(122)
a.quiver(x_pos,y_pos,x_dircn,y_direcn,scale = 7)
b.quiver(x_pos,y_pos,x_dir2,y_direcn,scale = 7)
plt.show()
```

```python
x = np.arange(0,2.2,0.1)
y = np.arange(0,2.2,0.1)
X,Y = np.meshgrid(x,y)
u = np.cos(X)*y
v = np.sin(y)*Y
fig = plt.figure(figsize = (5,5))
a=plt.subplot()
a.streamplot(X,Y,u,v,density = 1)
plt.show()
```
