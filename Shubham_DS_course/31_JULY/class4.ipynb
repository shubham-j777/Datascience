# Data Manipulation

```python
import numpy as np
# list
# 0
# 1 
# full
# linspace
# arange
# random
```

```python
import numpy as np
a = np.array([1,2,3,4])
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.zeros((3,6))
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.ones((3,6))
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.full(3,6)
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.linspace(3,6)
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.arange(3,6,1)
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
a = np.random.random((3,6))
print(a)
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

# Image import

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageOps

img = np.array(Image.open('image1.jpg'))
print(img)
plt.figure(figsize = (9,9))
plt.imshow(img)
```

```python
img_NB = img.copy()
# 0 = red
# 1 = green
# 2 = blue
print('''
green replaced with red    blue replaced with blue
blue  replaced with green     blue replaced with red''')
img_NB[0:2000,0:3000,1] = 0
img_NB[2000:3500,0:3000,2] = 3
img_NB[0:2000,3000:5000,2] = 1
img_NB[2000:3500,3000:5000,2] = 0

plt.figure(figsize = (9,9))
plt.imshow(img_NB)
```
